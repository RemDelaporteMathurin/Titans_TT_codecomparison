```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd



# Load data
COMS = np.loadtxt('TDS_COMSOL.txt')
fest = np.loadtxt('TDS_festim.txt')

# Create a new figure
plt.figure(2)

# Read table from CSV file
bb = pd.read_csv('Exp_OgorodnikovaJNM2003_fluence_1e23.csv', delimiter=';')
bb = pd.read_csv(f'Exp_OgorodnikovaJNM2003_fluence_1e23.csv', sep=';', names=['Temperature (K)', 'des flux (x1e19 D/m2/s)'])
cc = np.loadtxt('exp_T.txt')
dd = np.loadtxt('exp_flux.txt')
bb['des flux (x1e19 D/m2/s)'] *= 1E19

# Plot data
plt.plot(COMS[::10, 0], COMS[::10, 1], 'b', linewidth=1, label='mHIT')
plt.plot(fest[:, 0], fest[:, 1], 'r', linewidth=1, label='Festim')
plt.plot(bb['Temperature (K)'], bb['des flux (x1e19 D/m2/s)'] , label='experiment')
plt.plot(cc, dd, linewidth=1, label='MHIMS')
plt.xlabel('Temperature [K]')
plt.ylabel('Flux [atoms/m^2/s]')

# Plot more data


plt.legend(loc='upper right')
plt.xlim(301, 900)
plt.ylim(0, 1E19)

# Show the plot
plt.show()

```
