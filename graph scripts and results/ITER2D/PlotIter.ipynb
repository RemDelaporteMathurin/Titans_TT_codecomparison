```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np 

df=pd.read_csv('derived_quantities_solute.csv', delimiter=',')
x_column = 'x'
y_columns = ['y1', 'y2', 'y3', ..., 'yn']
df.columns.values[0]='x'
c_vec=['b','b','r','g','b','r','g']
c_vec=['#87CEEB','#87CEEB','#FF6B6B','#98FB98','#87CEEB','#FF6B6B','#98FB98']
c_vec2=['#87CEEB','#FF6B6B','#98FB98','y']
plt.plot([], [], linestyle='--', c='#666666',label='mHIT')  
plt.plot([], [], c='#666666',label='festim')  
plt.legend(loc='lower left', bbox_to_anchor=(0, 0.05),fontsize=12) 
ax = plt.subplot(111)
ax.spines[['right', 'top']].set_visible(False)
label_vec=['t','Tungsten','Copper','CuZr','Tungsten','Copper','CuZr']
for i,column in enumerate(df.iloc[:,:]):
    if column=='x':
        continue
    elif i<4:
        continue
    else:
         plt.plot(df['x'], df[column], c_vec[i], label=label_vec[i])
plt.plot(df['x'],df.iloc[:, 4:].sum(axis=1),'y', label='Total') 

filename_vec1=['cm']
filename_vec2=['W','Cu','CuCr','tot']
i=0
for el1 in filename_vec1:
    for el2 in filename_vec2:
        
        filename=el1+'_'+el2+'.txt'
        df=np.genfromtxt(filename, delimiter=None)
        plt.semilogy(df[:,0],df[:,1]/2,'--',color=c_vec2[i])
        i=i+1
       
plt.xscale('log')
plt.yscale('log')
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.xlabel('Time [s]')
plt.ylabel('Mobile H inventory [H/m]')
plt.legend()
plt.ylim([1E12,1E20])
plt.legend()
plt.title('Iter 2D- mobile T inventory')
plt.savefig('ITER2DSolute.png',dpi=450)

```
