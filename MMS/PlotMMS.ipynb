```python
import MMS_package as mms
from sympy import symbols, cos, pi
import matplotlib.pyplot as plt
plt.rcParams["text.latex.preamble"]

#plt.rcParams['text.usetex'] = False
x, y, t = symbols("x y t")
fun = (2 + cos(2 * pi * x) * cos(2 * pi * y)) * t
fun2 = (1+cos(2*pi*x)*cos(2*pi*y))*t

mms.plotfun(fun,K=10,n_lines=4,filename="MS10s.txt",mms=True)
mms.plotfun(fun2,K=10,n_lines=4,filename="Ver10s.txt",mms=False)
```
