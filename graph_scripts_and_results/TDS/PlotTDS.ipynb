```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd



# Load data
COMS = np.loadtxt('TDS_COMSOL.txt')
fest = np.loadtxt('TDS_festim.txt')

# Create a new figure
plt.figure(2)

# Read table from CSV file
bb=np.loadtxt('Exp_OgorodnikovaJNM2003_fluence_1e23.csv', delimiter=';', skiprows=1)
#bb = pd.read_csv('Exp_OgorodnikovaJNM2003_fluence_1e23.csv', delimiter=';')

#bb = pd.read_csv(f'Exp_OgorodnikovaJNM2003_fluence_1e23.csv', sep=';', names=['Temperature (K)', 'des flux (x1e19 D/m2/s)'])
#bb=bb.iloc[1::,:]
cc = np.loadtxt('exp_T.txt')
dd = np.loadtxt('exp_flux.txt')
#bb['des flux (x1e19 D/m2/s)'] *= 1E19
bb[:,1]=bb[:,1]*1E19
bb[:,0]=bb[:,0]
# Plot data
plt.plot(COMS[::10, 0], COMS[::10, 1], 'b', linewidth=1, label='mHIT')
plt.plot(fest[:, 0], fest[:, 1], 'r', linewidth=1, label='Festim')
#plt.plot(bb['Temperature (K)'], bb['des flux (x1e19 D/m2/s)'],'g--' , label='experiment')
plt.plot(bb[:,0], bb[:,1],'g--' , label='experiment')
plt.plot(cc, dd, linewidth=1, label='MHIMS')
plt.xlabel('Temperature [K]',size=16)
plt.ylabel(r'Flux $D/(m^2s)$',size=16)
plt.yticks(np.linspace(0, 1E19,5))

# Plot more data
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

plt.legend(loc='upper right',fontsize=16, frameon=False)
plt.xlim(301, 900)
plt.ylim(0, 1E19)

# Show the plot
plt.savefig('W_TDS_festim.png',dpi=450)

```
